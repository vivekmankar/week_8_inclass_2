```python
import pandas as pd 
import numpy as np 
import glob
all_data = pd.DataFrame() 
for f in glob.glob("0*.csv"):   
    df = pd.read_csv(f)    
    all_data = all_data.append(df,ignore_index=True) 
    all_data.describe()
```

```python
all_data.head()
```

```python
all_data["Confirmed"]=all_data["Confirmed"].fillna(0.0).astype(int)
```

```python
all_data["Deaths"]=all_data["Deaths"].fillna(0.0).astype(int)
```

```python
all_data["Recovered"]=all_data["Recovered"].fillna(0.0).astype(int)
```

```python
df = all_data.groupby('Country/Region').sum()
```
